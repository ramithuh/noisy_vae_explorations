```python
import pandas as pd
import plotly.express as px

import plotly.io as pio
pio.renderers.default = 'colab'
```

```python
# Read the data
df = pd.read_csv('uniform_seed.csv')

# Create box plot with points
fig = px.box(
    df,
    x='noise.range',
    y='avg_train_loss',
    points='all',
    labels={
        'noise.range': 'Noise_Range/2 (r/2)',
        'avg_train_loss': 'Average train loss'
    }
)

# Customize the layout with zero margins
fig.update_layout(
    width=700,
    height=500,
    showlegend=False,
    title=None,
    margin=dict(l=0, r=0, t=0, b=0),  # Remove all margins
    font=dict(
        size=20
    )
)

# Increase axis font size and add standoff
fig.update_xaxes(
    title_font=dict(size=26),
    tickfont=dict(size=22),
    title_standoff=25,  # Add space between x-axis title and ticks
    automargin=True     # Ensure margins adjust to accommodate the standoff
)

fig.update_yaxes(
    title_font=dict(size=26),
    tickfont=dict(size=22),
    title_standoff=25,  # Add space between y-axis title and ticks
    automargin=True     # Ensure margins adjust to accommodate the standoff
)

# Optional: Customize point and box appearance
fig.update_traces(
    marker=dict(color='blue', size=6),
    boxpoints='all',
    jitter=0.3,
    pointpos=-2,
    selector=dict(type='box')
)

# Show the plot
fig.show()

# Save with zero padding
fig.write_image("/home/ruh/Dropbox/Apps/Overleaf/10701_Project_GenAI_Noise_Analysis/generated_figs/uniform_seeds.pdf")
```
